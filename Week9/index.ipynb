```python
import hashlib

from statsmodels.stats.proportion import proportions_ztest
```

```python
def is_eligible(appointment):
    if appointment.patient.is_test_account:
        return False
    if appointment.patient.has_no_mobile_number():
        return False
    if appointment.type in ['video', 'phone']:
        return False
    return True
```

```python
def assign_group(patient_id, experiment_name):
    combined = f'{patient_id}_{experiment_name}'
    hash_hex = hashlib.md5(combined.encode()).hexdigest()
    group = int(hash_hex, 16) % 100

    return "control" if group < 50 else "treatment"


def analyse(ctrl_attend, ctrl_total, trt_attend, trt_total):
    rate_control = ctrl_attend / ctrl_total
    rate_treatment = trt_attend / trt_total
   
    lift = rate_treatment - rate_control

    z_stat, p_value = proportions_ztest([ctrl_attend, trt_attend], [ctrl_total, trt_total])
    return lift, z_stat, p_value
```
