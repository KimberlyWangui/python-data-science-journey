```python
# How to establish a connection to the database
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('bookstore.db')

print(conn)  # This will print the connection object, confirming that the connection was established successfully.
```

```python
# Using sql without pandas
cursor = conn.cursor()
cursor.execute("SELECT * FROM books")
results = cursor.fetchall()
print(results)
```

```python
# Using sql with pandas
# Read data from the database into a pandas DataFrame
pd.read_sql(
    "SELECT name FROM sqlite_master WHERE type='table';", # This query retrieves the names of all tables in the database
    conn
)
```

```python
# Selecting all from authors
pd.read_sql("SELECT * FROM authors;", conn)
```

```python
pd.read_sql("SELECT * FROM books;", conn)
```

```python
pd.read_sql("SELECT * FROM books WHERE published_year > 1990;", conn)
```

```python
pd.read_sql("SELECT * FROM customers;", conn)
```

```python
pd.read_sql("SELECT * FROM orders;", conn)
```

```python
pd.read_sql("SELECT author_id, name, country FROM authors;", conn)
```

```python
# rename columns using AS (aliasing): create temporary names for columns or tables to make the output more readable
pd.read_sql("SELECT title AS book_title, price AS price_usd FROM books;", conn)
```

```python
# Select author and name from authors table give name an alias of author_name
pd.read_sql("SELECT author_id, name AS author_name FROM authors;", conn)
```

```python
pd.read_sql("SELECT title, genre, price FROM books WHERE genre = 'Literary Fiction';", conn)
```

```python
pd.read_sql("SELECT title, genre, price FROM books WHERE genre = 'Literary Fiction' AND price < 12;", conn)
```

```python
pd.read_sql("SELECT * FROM books WHERE genre = 'Literary Fiction' OR price < 12;", conn)
```

```python
# IN and Between
pd.read_sql("SELECT title, genre, price FROM books WHERE genre IN ('Literary Fiction', 'Historical Fiction');", conn)
```

```python
pd.read_sql(
    """SELECT * FROM customers WHERE member_since BETWEEN 2019 AND 2020;""",
    conn
)
```

```python
pd.read_sql("SELECT * FROM customers WHERE name LIKE 'A%';", conn)
```

```python
pd.read_sql("""
SELECT * FROM books ORDER BY price ASC LIMIT 5;
""",
    conn
)
```

```python
pd.read_sql("""SELECT COUNT(*) AS total_books FROM books;""", conn)
```

```python
pd.read_sql("SELECT ROUND(AVG(price), 2) AS avg_price FROM books;", conn)
```

```python
pd.read_sql("SELECT SUM(quantity) AS total_quantity FROM orders;", conn)
```

```python
# GROUP BY and HAVING
pd.read_sql("""SELECT genre, COUNT(*) AS num_books FROM books GROUP BY genre HAVING num_books > 1;
""", conn)
```

```python
# select genre, title, avg(price), sum(price) of books group by genre, filter if price is greater than avg price
pd.read_sql("""SELECT genre, title, AVG(price) AS avg_price, SUM(price) AS total_price
    FROM books
    GROUP BY genre
    HAVING price > (SELECT AVG(price) FROM books);
    """, conn)
```
