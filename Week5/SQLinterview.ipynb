SELECT  name, salaries<br>
FROM employees<br>
ORDER BY salaries DESC<br>
LIMIT 5<br>
;

SELECT author_name, SUM(copies_sold) AS totalSold<br>
FROM Authors a<br>
JOIN Books b<br>
ON a.book_name = b.book_name<br>
GROUP BY author_name<br>
ORDER BY totalSold<br>
LIMIT 3<br>
;

SELECT c.name, COUNT(DISTINCT o.OrderDate) as NumOrderDates<br>
FROM (SELECT c.name, o.quantity FROM Customers c <br>
    JOIN Orders o ON c.orderNumber = o.OrderNumber<br>
    WHERE o.quantity > 1)<br>
WHERE NumOrderDates > 1;<br>

SELECT d.NAME, COUNT(e.ID) AS EmployeeCount<br>
FROM Department d<br>
LEFT JOIN Employee e<br>
ON e.DEPT_ID = d.DEPT_ID<br>
GROUP BY d.dept_id, d.name<br>
ORDER BY EmployeeCount DESC, d.NAME<br>
