# Bar Chart for Qualitative Data

```python
import matplotlib.pyplot as plt

# Data
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
counts = [10, 15, 7, 5]

# Create bar chart
plt.figure(figsize=(6, 4))
plt.bar(fruits, counts)
plt.xlabel('Fruits')
plt.ylabel('Counts')
plt.title('Favorite Fruits')

# Display the plot
plt.show()
```

# Histogram for Continuous Data

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random height data
heights = np.random.normal(170, 10, 100)

# Create histogram
plt.figure(figsize=(6, 4))
plt.hist(heights, bins=20, edgecolor='black')

# Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Student Heights')

# Display the plot
plt.show()
```

# Scatter Plot for Two Discrete Variables

```python
import matplotlib.pyplot as plt

# Data
hours_studied = [2, 3, 5, 1, 4, 2, 6, 3, 4, 5]
test_scores = [85, 92, 98, 78, 95, 88, 100, 90, 96, 99]

# Create scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(hours_studied, test_scores)

# Add labels and title
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')
plt.title('Hours Studied vs. Test Scores')

# Display the plot
plt.show()
```
