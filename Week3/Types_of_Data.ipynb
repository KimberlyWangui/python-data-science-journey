# **INTRODUCTION TO TYPES OF DATA AND DATA COLLECTION**

One very important skill in data science is understanding what kind of data you have. They can range from strings, integers, decimals, booleans etc.<br>
Choosing the right data type matters because it affects how you clean, analyze and interpret data.<br>
There are two broader categories:<br>
> i. Categorical (at times refered to as qualititive data)<br>
> ii Numerical (Quantitive)



## Categorical

Data that represent names, labels or groups<br>
There are two main types<br>
> **i. Nominal** -> Categorical data have no natural order i.e no level is greater than the other e.g gender, color, blood group<br>
sometimes they maybe stored as numbers but the numbers are only code representing categories.<br>
> **ii. Ordinal** -> Categorical data but the categories have a meaniful order but the difference between the categories is not measured. e.g Education levels, ratings<br>
> **iii.Binary categorical** -> Categories with only two types represented as booleans e.g yes and no,

## Numerical
Data that is measurable. Actual quantities<br>
Types of numerical:<br>
> **i. Discrete** -> Numerical data that comes from counting. e.g number of students (Whole numbers)<br>
> **ii. Continuos** -> Numerical data that comes from measurement e.g weight, height, stock price

There are number of methods you can use to clean text data.<br>
.replace(), .strip(), .upper(), .lower(), .capitalize(), .split()

## Tabular Data: The CSV file
CSV(comma seperated values): Stores data as plain text where each record(row) and each column is an attribute or feature.

name, age, city<br>
Francis, 34, Nairobi<br>
Antonia, 21, Moshi <br>
Trevor, 23, Nakuru<br>
Steve, 32, Kisumu

CSV in-built methods
1. csv.reader() -> it reads rows as lists
2. csv.DictReader() -> Reads rows as dictionaries using the column names
3. csv.writer() -> Writes rows in a csv
4. csv,DictWriter()- > Writes dictionaries into a csv

```python
import csv

with open("students.csv", "r") as file: # file = students.csv
    """
    students contains
    students = [
    
            ['name', ' age', ' city'],
            ['Francis', ' 34', ' Nairobi'],
            ['Antonia', ' 21', ' Moshi'],
            ['Trevor', ' 23', ' Nakuru'],
            ['Steve', ' 32', ' Kisumu']
            
            ]
    """
    students = csv.reader(file)

    for std in students:
        print(std)
```

## Data Collection and Sampling Techniques
Data Collection is the process of gathering data about a population.
Sampling is the process of taking a subset (sample) from a population data.

Population
----------
The full group one is interested in. <br>
Example: All students sitting KCSE exam in 2026 which is 1000000 students.

Sample
------
A smaller group taken from the population to be studied/analyzed <br>
Example: 1000 students slected from 40 schools across the country totalling to 40000

Sampling Bias
-------------
It's when the sample does not represent the population fairly which lead to misleading
conclusions. <br>
Example: You sample more gender than the other, sample more students from one region
neglecting the other regions.

Sampling matters since it helps in analyzing data. Good analysis should generalize the real world scenario which is only possible if the sample reflects the broader population well.

The main sampling techniques are:

i. Simple Random Sampling <br>
Each member of the population has on equal chance of being selected. <br>
Example: Randomly choose 100 students from a list of students.

It's normally applicable when the population is fairly known and you have access to the entire population (school list)

ii. Stratified Sampling <br>
You divide your population into important groups; stratas then you randomly sample each group. <br>
Example: Split customers by characteristics, e.g, agegroup, gender then sample each group.

Used when you want the important groups/sections of your population represented and there are known groups in the population.

iii. Cluster Sampling <br>
Your divide your population into clusters often identified by features such as location/convenience then you randomly choose some clusters and sample everyone/most people in the chosen clusters. <br>
Example: Choose a few schools from several provinces then survey students/part of students in those schools.

iv. Sequential Sampling <br>
You collect data step by step and decide whether to continue based on the results achieved so far. <br>
Example: Interviewing customers until a certain outcome/pattern emerges.
