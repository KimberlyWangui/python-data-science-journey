# Key Terms in Pyspark

**Apache Spark**: Distributed computing engine<br>
**Pyspark**: Python Interface for Apache spark<br>
**Cluster**: A group of nodes.<br>
**Cluster Manager** : It allocates the resources(cpu, gpu, memory) to worker nodes and coordinates the communication betweenthe driver program and the worker nodes. <br>
**Driver Program**: The main program that coordinates all spark jobs<br>
**Worker nodes**: The actual machine that perform the job<br>
**SparkContext**: Its the entry point for spark operation (unstructured data) <br>
**SparkSession**: Entry point to work with structured data, most specifically dataframes, sql, MLib<br>
**RDD**: It's a low level data structure in spark for collection of data. <br>
**DataFrame**: It's a combination of rows and columns - structured distributed tables. <br>
**MLib**: Spark's machine learning's library.

```python
from pyspark.sql import SparkSession
#from pyspark import SparkContext as sc
```

```python
spark = SparkSession.builder.appName('Practice').master("local[*]").getOrCreate()
```

```python
sc = spark.sparkContext
num = [2, 4, 7, 8, 9, 5, 11]
num_rdd = sc.parallelize(num)

def square(num):
    return num * num

def odd(num):
    return num % 2 != 0

def even(num):
    return num % 2 == 0


num_squares = num_rdd.map(square)
num_odd = num_rdd.filter(odd)
num_even = num_rdd.filter(even)

print(num_squares.collect())
print(num_odd.collect())
print(num_even.collect())
```

```python
# Shut down SparkContext
sc.stop()
```
