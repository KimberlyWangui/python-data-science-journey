### STATISTICAL INFERENCE FOR LINEAR REGRESSION

```python
import pandas as pd
"""
If a student studied for x hours, what will be their final score, and how uncertain is that score(prediction)?
"""

data = pd.DataFrame({
    "Hours_studied": [2, 2, 3, 4, 7, 4, 5, 6, 8, 9],
    "final_score": [50, 52, 60, 70, 90, 70, 77, 80, 85, 92]
})
data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.utils import resample
import numpy as np

X, y = data[["Hours_studied"]], data["final_score"]
model = LinearRegression()
model.fit(X, y)
model.predict(X) 
```
