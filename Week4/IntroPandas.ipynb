```python
passengers =  [
    {"name": "Alice", "age": 30, "survived": 1},
    {"name": "Bob", "age": 25, "survived": 0},
    {"name": "Charlie", "age": 35, "survived": 1},
    {"name": "David", "age": 28, "survived": 0},
    {"name": "Eve", "age": 22, "survived": 1}
    ]

# Average age of passengers
average_age = sum(passenger['age'] for passenger in passengers if passenger['age'] is not None) / len(passengers)
print(f"Average age of passengers: {average_age:.2f}")
```

```python
import pandas as pd
df = pd.DataFrame(passengers)
df
```

```python
average_age = df['age'].mean()
average_age
```
